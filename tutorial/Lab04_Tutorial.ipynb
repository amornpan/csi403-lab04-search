## Finding Keywords in Documents

**Time:** ~60 minutes

## Part 1: Check if Text Exists

Use `in` operator to check if a word exists in text.

```python
text = 'Rubella causes fever and rash'

# Check if 'fever' exists
if 'fever' in text:
    print('Found fever!')
else:
    print('Not found')

# Returns True/False
result = 'fever' in text
print(f'Result: {result}')
```

## Part 2: Case-Insensitive Search

```python
text = 'Rubella causes FEVER'

# This won't find it (case-sensitive)
print('fever' in text)  # False

# Convert both to lowercase
print('fever' in text.lower())  # True

# Or uppercase
print('FEVER' in text.upper())  # True
```

## Part 3: Count Occurrences

```python
text = 'The patient has fever. High fever is dangerous. Fever treatment is rest.'

# Count exact matches
count = text.count('fever')
print(f'fever count: {count}')  # 1 (case-sensitive!)

# Count case-insensitive
count = text.lower().count('fever')
print(f'fever count (case-insensitive): {count}')  # 3
```

## Part 4: Search Function

```python
def search(query, text):
    """Search for query in text (case-insensitive)."""
    return query.lower() in text.lower()

# Test
print(search('fever', 'Patient has FEVER'))  # True
print(search('cough', 'Patient has fever'))  # False
```

## Part 5: Search Multiple Documents

```python
def search_documents(query, documents):
    """Find all documents containing query."""
    results = []
    for doc in documents:
        if query.lower() in doc.lower():
            results.append(doc)
    return results

# Test
docs = [
    'Rubella causes fever and rash',
    'Cholera causes severe diarrhea',
    'Dengue fever is mosquito-borne'
]

results = search_documents('fever', docs)
print(f'Found {len(results)} documents:')
for r in results:
    print(f'  - {r}')
```

## Summary

| Concept | Code | Description |
|---------|------|-------------|
| Check exists | `'word' in text` | Returns True/False |
| Case-insensitive | `text.lower()` | Convert to lowercase |
| Count | `text.count('word')` | Count occurrences |
| Search docs | Loop + in | Find matching docs |
